Concatenation

```python
A=['a0','a1','a2','a3']
B=['b0','b1','b2','b3']
C=['c0','c1','c2','c3']
D=['d0','d1','d2','d3']
list_label=['A','B','C','D']
columns=[A,B,C,D]
zipped=list(zip(list_label,columns))
new_data=dict(zipped)
data1=pd.DataFrame(new_data)
```

```python
data1
```

```python
A=['a0','a5','a6','a7']
B=['b1','b5','b6','b7']
C=['c2','c5','c6','c7']
D=['d3','d5','d6','d7']
list_label=['A','B','C','D']
columns=[A,B,C,D]
zipped=list(zip(list_label,columns))
new_data=dict(zipped)
data2=pd.DataFrame(new_data)
```

```python
A=['a8','a9','a10','a11']
B=['b8','b9','b10','b11']
C=['c8','c9','c10','c11']
D=['d8','d9','d10','d11']
list_label=['A','B','C','D']
columns=[A,B,C,D]
zipped=list(zip(list_label,columns))
new_data=dict(zipped)
data3=pd.DataFrame(new_data)
```

```python
data2
```

```python
row_concat=pd.concat([data1,data2,data3])
```

```python
row_concat
```

```python
data1.merge(data2, left_on='A', right_on='A')
```
